### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df1 = df[df['Medal']=='Gold']
```

```python
df1
```

```python
x = df1.drop(['Year','City','Discipline','Athlete','Country','Gender','Event'], axis = 1)
```

```python
x
```

```python
y = x['Sport'].value_counts().head()
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.plot(y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Sports having most number of Gold medals so far')
plt.show()
```
